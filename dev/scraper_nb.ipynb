```python
events = soup.find_all('div', class_='siteorigin-widget-tinymce')
events_list = []

for event in events:
    event_dict = {}
    try:
        day = event.find('h3').text.split(' ')[1]
        day = datetime.datetime.strptime(f'{day} {datetime.datetime.now().month} {datetime.datetime.now().year}', '%d %m %Y')
    except:
        continue
    event_dict["title"] = event.find('h4').text
    event_dict["location"] = 'Fanfulla 5/A Circolo Arci'
    time = event.find('span', class_='_4n-j fsl').text.split('dalle ore ')[1]
    time = datetime.datetime.strptime(time, '%H').strftime('%H:%M')
    event_dict["date_and_time"] = day.replace(hour=int(time.split(':')[0]), minute=int(time.split(':')[1]))
    event_dict["url"] = event.find_all('a')[-1]['href']
    events_list.append(event_dict)
    print(events_list)

df = pd.DataFrame(events_list)
# df['date'] = df['date_and_time'].apply(lambda x: x.date())
# df['time'] = df['date_and_time'].apply(lambda x: x.time().strftime('%H:%M'))
# df['year'] = df['date_and_time'].apply(lambda x: x.year)
# df['month'] = df['date_and_time'].apply(lambda x: x.month)
# df['day'] = df['date_and_time'].apply(lambda x: x.day)
# df.drop('date_and_time', axis=1, inplace=True)
df = df[['date_and_time','title', 'location', 'url']]
```
